```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tdastro.obstable.opsim import OpSim
from tqdm.auto import tqdm # For progress bar

# --- 1. Initial Data Loading ---
# Load the OpSim database, which contains the simulated observation schedule for LSST.
OPSIM_DB_PATH = "/Users/sidchaini/miniconda3/envs/tdastro/lib/python3.12/data/opsim/baseline_v5.0.0_10yrs.db"
ops_data = OpSim.from_db(OPSIM_DB_PATH)

print(f"Loaded an opsim database with {len(ops_data)} entries.")
print(f"Columns: {ops_data.columns}")

# Load the theoretical light curve data and metadata.
lc = pd.read_csv("lc.csv", sep=" ").set_index("SNID")
meta = pd.read_csv("meta.csv", sep=" ").set_index("SNID")

# Ensure the filter column 'FLT' is lowercase to match OpSim 'band' column.
lc["FLT"] = lc["FLT"].str.lower()

# --- 2. Separate Survey Strategies ---
# The OpSim database contains different survey programs. We are interested in the main
# survey, not the Deep Drilling Fields (DDF), so we separate them.
ddf_idx = ops_data["science_program"][ops_data["science_program"] == "DD"].index
ddf_table = ops_data._table.iloc[ddf_idx]

nonddf_idx = ops_data["science_program"][ops_data["science_program"] != "DD"].index
nonddf_table = ops_data._table.iloc[nonddf_idx]

# Define a color mapping for plotting LSST filters.
lsstcolorplot = {
    "u": "#785EF0",
    "g": "#7DDEB2",
    "r": "#DC267F",
    "i": "#FE6100",
    "z": "#FFB000",
    "y": "#5F3331",
}
```

```python
# --- 3. Select a Sample Light Curve and a Sky Location ---

# We'll use the non-DDF table for our main survey simulation.
table = nonddf_table

# Choose a random supernova from our metadata table.
meta_idx = meta.sample(n=1).index[0]
print(f"Chose SNID {meta_idx} ({meta.loc[meta_idx, 'SIM_TYPE_NAME']})")

# Get the full, original light curve for this supernova.
# We use .copy() to avoid SettingWithCopyWarning later.
original_lc = lc.loc[meta_idx].copy()

# --- 4. Simulate Placing the Supernova into the LSST Survey ---

# Pick a random observation from the OpSim survey to act as our supernova's location (RA, Dec).
# This simulates the supernova appearing at a random place in the LSST sky.
obs_location = table.sample(n=1).iloc[0]
obs_ra = obs_location["ra"]
obs_dec = obs_location["dec"]

# We shift the supernova's explosion time to coincide with the LSST survey timeline.
# A random offset is added to vary the light curve's peak relative to the observation cadence.
time_offset = obs_location["time"] - original_lc["MJD"].min() + np.random.randint(low=-100, high=200)
original_lc["LSSTDATE"] = original_lc["MJD"] + time_offset

# --- 5. Crop the Light Curve Based on Actual LSST Observations ---

# Get all actual LSST observations from the OpSim database that fall
# near our chosen location and within the supernova's simulated timespan.
lsst_observations = pd.DataFrame(
    ops_data.get_observations(
        query_ra=obs_ra,
        query_dec=obs_dec,
        cols=["time", "band"],
        t_min=original_lc["LSSTDATE"].min(),
        t_max=original_lc["LSSTDATE"].max()
    )
)

# If there are no LSST observations in this window, we can't crop.
if lsst_observations.empty:
    print("No LSST observations found for this location and time. Try another random sample.")
else:
    # Prepare both dataframes for merging. They need a common time column and must be sorted.
    lsst_observations.sort_values('time', inplace=True)
    
    # Rename columns for clarity before merging.
    lc_to_merge = original_lc.rename(columns={'LSSTDATE': 'time', 'FLT': 'band'})
    lc_to_merge.sort_values('time', inplace=True)

    # This is the key step: For each actual LSST observation, find the NEAREST point
    # in our theoretical light curve (in time and for that specific filter/band).
    # 'tolerance' ensures we only match points that are reasonably close in time.
    # We use the max time difference from the original LC as a sensible tolerance.
    cropped_lc = pd.merge_asof(
        lsst_observations,
        lc_to_merge,
        on='time',
        by='band',
        direction='nearest',
        tolerance=original_lc["MJD"].diff().max()
    )

    # The merge may create rows with NaNs if no matching theoretical point was found within the tolerance.
    # We drop these to get our final, observed light curve.
    cropped_lc.dropna(inplace=True)
    
    # Rename columns back to the original format for consistency.
    cropped_lc = cropped_lc.rename(columns={'band': 'FLT'})


# --- 6. Visualize the Result ---

# Add a 'MJD_DIFF' column to both light curves to align them at day 0 for plotting.
min_date =  original_lc["MJD"].min()
original_lc["MJD_DIFF"] = original_lc["MJD"] - min_date
cropped_lc["MJD_DIFF"] = cropped_lc["MJD"] - min_date

plt.figure(figsize=(10, 6))
plt.title(f"Original vs. LSST-Cropped Light Curve for SNID {meta_idx}")
plt.xlabel("Days Since First Observation")
plt.ylabel("FLUXCAL")

for band_filter in "ugrizy":
    # Plot the full, original light curve data points (small and semi-transparent)
    band_df_orig = original_lc[original_lc["FLT"] == band_filter]
    plt.scatter(
        band_df_orig["MJD_DIFF"], band_df_orig["FLUXCAL"],
        color=lsstcolorplot[band_filter], s=5, alpha=0.5
    )
    
    # Plot the cropped, "observed" data points (larger and opaque)
    band_df_crop = cropped_lc[cropped_lc["FLT"] == band_filter]
    plt.scatter(
        band_df_crop["MJD_DIFF"], band_df_crop["FLUXCAL"],
        color=lsstcolorplot[band_filter], s=50, label=f'{band_filter}-band', edgecolor='black'
    )

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
def crop_light_curve(lc_df: pd.DataFrame, opsim_data: OpSim, survey_table: pd.DataFrame) -> pd.DataFrame:
    """
    Simulates observing a given light curve with the LSST survey.

    Args:
        lc_df (pd.DataFrame): DataFrame for a single, well-sampled light curve.
                              Must contain 'MJD' and 'FLT' columns.
        opsim_data (OpSim): The loaded OpSim database object from tdastro.
        survey_table (pd.DataFrame): The OpSim observation table to sample from (e.g., non-DDF).

    Returns:
        pd.DataFrame: A new DataFrame representing the "observed" or cropped light curve.
                      Returns an empty DataFrame if no observations were found.
    """
    # Use .copy() to ensure the original DataFrame is not modified.
    input_lc = lc_df.copy()
    
    # 1. Pick a random sky location and time from the survey.
    obs_location = survey_table.sample(n=1).iloc[0]
    obs_ra = obs_location["ra"]
    obs_dec = obs_location["dec"]
    
    # 2. Shift the light curve's time to align with the survey time.
    time_offset = obs_location["time"] - input_lc["MJD"].min() + np.random.randint(low=-100, high=200)
    input_lc["LSSTDATE"] = input_lc["MJD"] + time_offset

    # 3. Find all actual LSST observations for that location and time.
    lsst_observations = pd.DataFrame(
        opsim_data.get_observations(
            query_ra=obs_ra,
            query_dec=obs_dec,
            cols=["time", "band"],
            t_min=input_lc["LSSTDATE"].min(),
            t_max=input_lc["LSSTDATE"].max()
        )
    )

    if lsst_observations.empty:
        return pd.DataFrame() # Return an empty frame if no observations

    # 4. Match LSST observations to the theoretical light curve points.
    lsst_observations.sort_values('time', inplace=True)
    lc_to_merge = input_lc.rename(columns={'LSSTDATE': 'time', 'FLT': 'band'})
    lc_to_merge.sort_values('time', inplace=True)
    
    max_cadence = lc_to_merge["time"].diff().max()
    
    # Handle cases where max_cadence might be NaN (if only one point)
    if pd.isna(max_cadence):
        max_cadence = 2.0 # Default to 2 days if cadence cannot be calculated

    cropped_df = pd.merge_asof(
        lsst_observations,
        lc_to_merge,
        on='time',
        by='band',
        direction='nearest',
        tolerance=max_cadence
    )
    
    # 5. Clean up and return the result.
    cropped_df.dropna(inplace=True)
    
    # Restore original column names and remove temporary ones.
    cropped_df = cropped_df.rename(columns={'band': 'FLT'})
    cropped_df = cropped_df.drop(columns=['time'])

    return cropped_df

# --- Test the function with our previous example ---
original_lc_for_func = lc.loc[meta_idx]
test_crop = crop_light_curve(original_lc_for_func, ops_data, nonddf_table)
print(f"Function produced a cropped light curve with {len(test_crop)} data points.")
test_crop.head()
```

```python
# Get a list of all unique supernova IDs.
all_snids = lc.index.unique()

# Create a list to hold the resulting cropped light curve DataFrames.
all_cropped_lcs = []

# Loop through each SNID, applying the cropping function.
# tqdm adds a helpful progress bar.
for snid in tqdm(all_snids, desc="Cropping all light curves"):
    # Get the original light curve for this SNID.
    single_lc = lc.loc[snid]
    
    # Run the cropping simulation.
    cropped_lc = crop_light_curve(single_lc, ops_data, nonddf_table)
    while len(cropped_lc)==0:
        cropped_lc = crop_light_curve(single_lc, ops_data, nonddf_table)

    cropped_lc["snid"] = (np.ones(shape=len(cropped_lc))*snid).astype("int")
    
    # If the result is not empty, add it to our list.
    if not cropped_lc.empty:
        all_cropped_lcs.append(cropped_lc)

# Concatenate all the individual cropped light curves into a single master DataFrame.
final_cropped_dataset = pd.concat(all_cropped_lcs)

print(f"\nFinished processing.")
print(f"Successfully cropped {final_cropped_dataset["snid"].nunique()} out of {len(all_snids)} light curves.")
print(f"Total number of observed points: {len(final_cropped_dataset)}")

final_cropped_dataset.head()
```
